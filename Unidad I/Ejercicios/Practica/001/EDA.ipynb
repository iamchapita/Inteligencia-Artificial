### Ejercicio cuatro: Segun el ejercicio tres, que tanto aumento o disminuyo el precio de cada item-variety 

Ejemplo de como se debe de mostrar su respuesta

```python
pd.DataFrame({
    "item" : ['Manzana','Manzana','Manzana']
    , "variety" : ['Gala','Gala','Gala']
    , "year" : [2020,2021,2022]
    , "last_price" :  [2.7,2.5,3.2]
    , "delta" : [0, (2.5-2.7), (3.2-2.5)]
})
```

Segun nuestro ejemplo dummy para 2021 estuvo 0.2 centavos mas barato y en 2022 estuvo 0.7 centavos caro

```python
deltaPrice = vdf.copy()
```

```python
deltaPrice['count'] = deltaPrice.groupby(['category', 'item', 'variety', 'year'], as_index=False).cumcount()
```

```python
deltaPrice = deltaPrice.loc[deltaPrice['count'] == 0 , ['item', 'variety', 'year','price']]
```

```python
deltaPrice['prevPrice'] = deltaPrice['price'].shift(1)
```

```python
deltaPrice['delta'] = deltaPrice.apply(lambda r: r['price'] - r['prevPrice'], axis=1)
```

```python
deltaPrice['delta'] = deltaPrice['delta'].fillna(0)
```

```python
deltaPrice.loc[:,['item', 'variety', 'year', 'price', 'delta']]
```

### Ejercicio seis: muestre por category,item,variety que tanto % ha aumentado su precio

Ejemplo: si para las manzanas gala el primer precio que se registro fue 2.00 y el ultimo precio que se registro es 4.00 este aumento un 100%

```python
pd.DataFrame({
    "category": ["fruta","fruta"]
    ,"item" : ['Manzana','Manzana']
    , "variety" : ['Gala','egremont_russet']
    , "percentage-change-between-first-and-last-price" : [200,150]       
})
```

```python
vdf = vdf.sort_values(['category', 'item', 'variety', 'year', 'month', 'day'])
```

```python
percentage = vdf.copy()
```

```python
percentage['cumCount'] = percentage.groupby(['category', 'item', 'variety'], as_index=False).cumcount()
```

```python
left = percentage.loc[(percentage['cumCount'] == 0) ,:]
```

```python
left
```

```python
right = percentage.groupby(['category', 'item', 'variety'], as_index=False).max()
```

```python
right
```

```python
percentage = pd.merge(left, right, how='inner', left_on=['category', 'item', 'variety'], right_on=['category', 'item', 'variety'])
```

```python
percentage = percentage.loc[:, ['category', 'item', 'variety', 'price_x', 'price_y']]
```

```python
percentage['percentage-changed-between-firstyear-and-lastyear'] = percentage.apply(lambda r: np.round(((r['price_y'] / r['price_x'])*100), decimals=1), axis=1)
```

```python
percentage = percentage.loc[:,['category', 'item', 'variety', 'percentage-changed-between-firstyear-and-lastyear']]
```

```python
percentage
```

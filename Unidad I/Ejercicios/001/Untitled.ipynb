Se obtiene una tabla con la cantidad de valores nulos de cada columna con su respectivo porcentaje

```python
null_summary = pd.DataFrame()
```

```python
for c in salaries.columns:
    r = pd.DataFrame({
        "Column": [c], 
        "NaFound": [ np.sum(pd.isnull([salaries[c]])) ],
        "Percentage": [ np.sum(pd.isnull([salaries[c]])) / salaries.shape[0] ] 
    })
    null_summary = pd.concat([ null_summary, r] )
```

## Obtener el nÃºmero y porcentaje de datos nulos de cada columna por aÃ±o

```python
a = salaries.loc[: , ['work_year']].drop_duplicates()
```

```python
null_summary = pd.DataFrame()
```

```python
for a in salaries.loc[:, 'work_year'].drop_duplicates():
    salarybyyear = salaries.loc[salaries['work_year'] == a, :]
    for c in salaries.columns:
        r = pd.DataFrame({
            "column": [c],
            "NaFound": [ np.sum(pd.isnull( salarybyyear[c] )) ],
            "Percentage": [ np.sum(pd.isnull( salarybyyear[c] )) / salarybyyear.shape[0] ],
            "work_year": [a]
        })
        
        null_summary = pd.concat([null_summary, r])
```

```python
null_summary.shape
```

```python
def getNullSummaryByColumn(dataFrame, column):
    null_summary = pd.DataFrame()
    for a in dataFrame.loc[:, column].drop_duplicates():
        subSet = dataFrame.loc[dataFrame[column] == a, :]
        for c in dataFrame.columns:
            r = pd.DataFrame({
                "column": [c],
                "NaFound": [ np.sum(pd.isnull( subSet[c] )) ],
                "Percentage": [ np.sum(pd.isnull( subSet[c] )) / subSet.shape[0] ],
                "work_year": [a]
            })

            null_summary = pd.concat([null_summary, r])
    return null_summary
```

```python
getNullSummaryByColumn(salaries, 'work_year')
```
